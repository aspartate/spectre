```python
from Spectre import Spectre
import numpy as np

# Note: I haven't 3D-printed all these yet, so I don't know if they'll actually tile together properly.
# May need to adjust the clearance and roundness parameters depending on your printer/laser cutter.

# Vanilla Spectre
spectre = Spectre(side_length = 100, round = 0.01, clearance = 0.02)
spectre.save('spectre_base.svg')

# Hollow Spectre
spectre = Spectre(side_length = 100, round = 0.01, clearance = 0.02)
spectre.make_hollow(edge_width = 20)
spectre.save('spectre_hollow.svg')

# Blocky Spectre
spectre = Spectre(side_length = 100, round = 0.05, clearance = 0.01)
spectre.set_profile([(0, 0), (0.5, 0.5), (1, 0)])
spectre.make_hollow(edge_width = 25)
spectre.save('spectre_blocky.svg')

# Wiggly Spectre
spectre = Spectre(side_length = 100, round = 0.01, clearance = 0.02)
r = 0.333333 / 2
x = np.linspace(0, 2*r, 200)
y = np.sqrt(r**2 - (x - r)**2)
profile_points = list(zip(x, y)) + list(zip(x + 2 * r, -y)) + list(zip(x + 4 * r, y))
spectre.set_profile(profile_points)
spectre.make_hollow(edge_width = 25)
spectre.save('spectre_wiggly.svg')
```
